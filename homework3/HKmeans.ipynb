# Section 2: K-means clustering from scratch

```python
def kmeans(data, k):
    threshold = 0.0000000001
    data = data.to_numpy()
    centers = {}
    
    #initialize the centroids, the first 'k' elements in the dataset will be our initial centroids
    for i in range(k):
        centers[i] = data[i]
        
    converged = False
    loop_counter = 0
        
    while not converged:
        classes = {}
        for i in range(k):
            classes[i] = []

        #find the distance between the point and cluster; choose the nearest centroid
        for features in data:
            distances = [np.linalg.norm(features - centers[i]) for i in centers]
            classification = distances.index(min(distances))
            classes[classification].append(features)
            
        previous_centers = dict(centers)

        #average the cluster datapoints to re-calculate the centroids
        for classification in classes:
            centers[classification] = np.average(classes[classification], axis = 0)
        
        converged = True
        for i in centers:
            prev_center = previous_centers[i]
            current_center = centers[i]

            if np.sum((current_center - prev_center)/prev_center * 100.0) > threshold:
                converged = False
                
        loop_counter += 1
    
    predictions = []
    wcss = 0
    for row in data:
        distances = [np.linalg.norm(row - centers[i]) for i in centers]
        pred = distances.index(min(distances))
        predictions.append(pred)
        wcss += np.linalg.norm(row - centers[pred])
        
    return centers, predictions, wcss, loop_counter
```

```python
X = data.drop('Country', 1)
```

```python
k_s = [5, 10, 15, 20]

all_wcss = []
for k in k_s:
    centers, pred, wcss, loop_counter = kmeans(X, k)
    all_wcss.append(wcss)
    print("For k:", k, ", count of loops:", loop_counter)
    
    #===================    
    # distribution of clusters
    pred = np.asarray(pred)
    plt.clf()
    plt.xlim([min(pred)-1, max(pred)+1])
    plt.hist(pred, bins='auto', edgecolor='black', linewidth=1.2)
    plt.xlabel('Clusters (k='+str(k)+')')
    plt.ylabel('Number of countries in the clusters')
    plt.show()
    # plt.savefig('k_'+str(k)+'_clusters.eps', format='eps')

# plot
plt.plot(k_s, all_wcss, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
# plt.savefig('manual_discreet.eps', format='eps')
```

# Section 3: K-means clustering using Scikit-learn

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=10, init='k-means++',n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(X)
```

```python
# plot the clusters based on 1st 2 features only
plt.scatter(X.iloc[:,0], X.iloc[:,1])

# plot the centroids
plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
# calculate distortion for a range of number of cluster
k_s = [5, 10, 15, 20]

all_wcss = []
for k in k_s:
    km = KMeans(
        n_clusters=k, init='k-means++',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(X)
    all_wcss.append(km.inertia_)
    print("For k:", k, ", count of loops:", km.n_iter_)

# plot
plt.plot(k_s, all_wcss, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
# plt.show()
plt.savefig('scikit_discreet.eps', format='eps')
```

```python
# calculate distortion for a range of number of cluster
all_wcss = []
for k in range(1,51):
    km = KMeans(
        n_clusters=k, init='k-means++',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(X)
    all_wcss.append(km.inertia_)
    print("For k:", k, ", count of loops:", km.n_iter_)

# plot
plt.plot(range(1,51), all_wcss, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
# plt.show()
plt.savefig('scikit_continuous.eps', format='eps')
```
